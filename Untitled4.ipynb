```python
!pip install -q huggingface_hub

from huggingface_hub import InferenceClient

# अपना टोकन यहाँ डालें
HF_TOKEN = "your hf token here"

try:
    client = InferenceClient(token=HF_TOKEN.strip())

    print("✅ कनेक्ट हो रहा है...\n")

    # Qwen 2.5 72B ओपन-सोर्स मॉडल
    response = client.chat_completion(
        model="Qwen/Qwen2.5-72B-Instruct",
        messages=[
            {"role": "user", "content": "नमस्ते! क्या सब सही काम कर रहा है?"}
        ],
        max_tokens=150
    )

    print("----------------------------------")
    print("AI का जवाब:", response.choices[0].message.content)
    print("----------------------------------")
    print("🎉 बधाई हो! मॉडल पूरी तरह से काम कर रहा है!")

except Exception as e:
    print(f"❌ Error: {e}")
```
